```python
import numpy as np
from scipy.stats import norm
```

## Gassuan PDF and CDF

```python
norm.pdf(0, loc=5, scale=10)
```

```python
arr = np.random.randn(10)
norm.pdf(arr)
```

During most of the time, we need to calculate the joint probability of data samples, instead of doing probability multiplication, we prefer to calculate **the sum of log(probability)**

```python
norm.logpdf(arr)
```

```python
norm.cdf(arr)
```

```python
norm.logcdf(arr)
```

## Sanpling from a 1D Gaussian distribution 

```python
arr = np.random.randn(10000)
```

```python
import matplotlib.pyplot as plt

plt.hist(arr, bins=100)
plt.show()
```

```python
arr2 = 10 * np.random.randn(10000) + 5

# non-standard Gaussian distribution
plt.hist(arr2, bins=100)
plt.show()
```

```python
arr = np.random.randn(10000, 2)

plt.scatter(arr[:, 0], arr[:, 1])
plt.show()
```

```python
arr[:, 1] = 100 * arr[:, 1] + 2

plt.scatter(arr[:, 0], arr[:, 1])
plt.show()
# scale changes in y-axis
# plt.axis('equal')
```

## Sampling from a General Multivariate Normal

```python
from scipy.stats import multivariate_normal as mvn
```

```python
cov = np.array([[1, 0.8], [0.8, 3]])
mu = np.array([0, 2])

arr = mvn.rvs(mean=mu, cov=cov, size=1000)
# arr = np.random.multivariate_normal(mean=mu, cov=cov, size=1000)
```

```python
plt.axis('equal')
plt.scatter(arr[:, 0], arr[:, 1])
plt.show()
```
